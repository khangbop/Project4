

***2.   Shanghai Data***



```python
df_shanghaiData['world_rank'] = df_shanghaiData['world_rank'].astype('string')
df_shanghaiData['university_name'] = df_shanghaiData['university_name'].astype('string')
df_shanghaiData['national_rank'] = df_shanghaiData['national_rank'].astype('string')
```



***3.***   ***Times Data***

```python
df_timesData = pd.read_csv('/content/drive/MyDrive/DA/FinalProject2/Dataset/timesData.csv')
```

```python
df_timesData
```

```python
df_timesData.info()
```

```python
df_timesData = df_timesData.replace('-',np.nan)
df_timesData = df_timesData.replace(' ',np.nan)
```

```python
df_timesData['world_rank'] = df_timesData['world_rank'].astype('string')
df_timesData['university_name'] = df_timesData['university_name'].astype('string')
df_timesData['country'] = df_timesData['country'].astype('string')
df_timesData['female_male_ratio'] = df_timesData['female_male_ratio'].astype('string')
df_timesData['international_students'] = df_timesData['international_students'].astype('string')
```

```python
df_timesData['international'] = df_timesData['international'].astype('float')
df_timesData['total_score'] = df_timesData['total_score'].astype('float')
df_timesData['income'] = df_timesData['income'].astype('float')

df_timesData['num_students'] = df_timesData['num_students'].str.replace(',','')
df_timesData['num_students'] = df_timesData['num_students'].astype('string').astype('float')
```
